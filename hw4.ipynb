```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread(r'C:\Users\25758\Pictures\1104picture.jpg')

gray_image = image.mean(axis=2)

U, S, VT = np.linalg.svd(gray_image)

def compress_and_show(image, singular_values_ratio):
    k = int(singular_values_ratio * len(S))
    reconst_img = np.dot(U[:, :k], np.dot(np.diag(S[:k]), VT[:k, :]))
    plt.imshow(reconst_img, cmap='gray')
    plt.title(f'{singular_values_ratio*100:.0f}% singular values')
    plt.axis('off')
    plt.show()

# 
for ratio in [0.01, 0.02, 0.1, 0.5]:
    compress_and_show(gray_image, ratio)
```
